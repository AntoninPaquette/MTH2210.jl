# Question 1

Créez un vecteur allant de 1 à 23. Ensuite, créez un vecteur contenant le carré du vecteur précédent et afficher dans un tableau les résultats.

```julia
a = 1:2:23
b = a .^ 2

@printf("n    a                       b\n");
for t=1:length(a)
    @printf("%2d   %16.15e   %16.15e\n",t,a[t],b[t])
end
```

# Question 2

Affichez sur le même graphique les fonctions $f$ et $g$ suivantes
\begin{align*}
    f(x) &= sin(x)^2\\
    g(x) &= 0.2x + 1
\end{align*}
pour $x\in [-2\pi,2\pi]$.

```julia
f(x) = sin(x)^2
g(x) = 0.2*x + 1

x = range(-2*pi,2*pi,length=250)
y1 = f.(x)
y2 = g.(x)

plot(x,y1,color="black")
plot!(x,y2,color="red",xlabel="x",title="Fonction f et g")
```
